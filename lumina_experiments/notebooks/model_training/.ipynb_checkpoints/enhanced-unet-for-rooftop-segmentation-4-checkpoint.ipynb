```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class ResidualBlock(nn.Module):
    """
    Novelty 1: Residual Block
    Replaces standard double conv. Helps in training deeper networks 
    and preserving feature identity.
    """
    def __init__(self, in_c, out_c):
        super().__init__()
        self.conv1 = nn.Conv2d(in_c, out_c, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(out_c)
        self.conv2 = nn.Conv2d(out_c, out_c, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(out_c)
        
        # Shortcut connection to match dimensions
        self.shortcut = nn.Sequential()
        if in_c != out_c:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_c, out_c, kernel_size=1, padding=0),
                nn.BatchNorm2d(out_c)
            )

    def forward(self, x):
        shortcut = self.shortcut(x)
        x = F.relu(self.bn1(self.conv1(x)))
        x = self.bn2(self.conv2(x))
        x += shortcut  # The "Residual" connection
        return F.relu(x)

class AttentionGate(nn.Module):
    """
    Novelty 2: Attention Gate
    Filters the skip connection features. It uses the coarser signal 
    from the decoder (g) to highlight relevant regions in the encoder features (x).
    """
    def __init__(self, in_c, gate_c, inter_c):
        super().__init__()
        # W_g: Transform gating signal
        self.Wg = nn.Sequential(
            nn.Conv2d(gate_c, inter_c, kernel_size=1, padding=0),
            nn.BatchNorm2d(inter_c)
        )
        # W_x: Transform skip connection
        self.Wx = nn.Sequential(
            nn.Conv2d(in_c, inter_c, kernel_size=1, padding=0),
            nn.BatchNorm2d(inter_c)
        )
        # Psi: Combine and activate
        self.psi = nn.Sequential(
            nn.Conv2d(inter_c, 1, kernel_size=1, padding=0),
            nn.BatchNorm2d(1),
            nn.Sigmoid()
        )
        
    def forward(self, x, g):
        g1 = self.Wg(g)
        x1 = self.Wx(x)
        psi = F.relu(g1 + x1)
        psi = self.psi(psi)
        return x * psi  # Scale the skip connection by the attention map

class ResAttUNet(nn.Module):
    def __init__(self, in_channels=3, out_channels=1):
        super().__init__()
        
        # --- Encoder (Downsampling) ---
        self.e1 = ResidualBlock(in_channels, 64)
        self.e2 = ResidualBlock(64, 128)
        self.e3 = ResidualBlock(128, 256)
        self.e4 = ResidualBlock(256, 512)
        
        self.pool = nn.MaxPool2d(2, 2)
        
        # --- Bridge ---
        self.b = ResidualBlock(512, 1024)
        
        # --- Decoder (Upsampling + Attention) ---
        
        # Up 1
        self.up1 = nn.Upsample(scale_factor=2, mode="bilinear", align_corners=True)
        self.att1 = AttentionGate(in_c=512, gate_c=1024, inter_c=256)
        self.d1 = ResidualBlock(1024 + 512, 512)
        
        # Up 2
        self.up2 = nn.Upsample(scale_factor=2, mode="bilinear", align_corners=True)
        self.att2 = AttentionGate(in_c=256, gate_c=512, inter_c=128)
        self.d2 = ResidualBlock(512 + 256, 256)
        
        # [NEW] Deep Supervision Output 2 (from 256 channels)
        self.ds_out3 = nn.Conv2d(256, out_channels, kernel_size=1) 
        
        # Up 3
        self.up3 = nn.Upsample(scale_factor=2, mode="bilinear", align_corners=True)
        self.att3 = AttentionGate(in_c=128, gate_c=256, inter_c=64)
        self.d3 = ResidualBlock(256 + 128, 128)

        # [NEW] Deep Supervision Output 3 (from 128 channels)
        self.ds_out2 = nn.Conv2d(128, out_channels, kernel_size=1)
        
        # Up 4
        self.up4 = nn.Upsample(scale_factor=2, mode="bilinear", align_corners=True)
        self.att4 = AttentionGate(in_c=64, gate_c=128, inter_c=32)
        self.d4 = ResidualBlock(128 + 64, 64)
        
        # Output (Final)
        self.out = nn.Conv2d(64, out_channels, kernel_size=1)
        
    def forward(self, x):
        # Encoder
        x1 = self.e1(x)
        p1 = self.pool(x1)
        
        x2 = self.e2(p1)
        p2 = self.pool(x2)
        
        x3 = self.e3(p2)
        p3 = self.pool(x3)
        
        x4 = self.e4(p3)
        p4 = self.pool(x4)
        
        # Bridge
        b = self.b(p4)
        
        # Decoder
        
        # Block 1
        d1 = self.up1(b)
        x4 = self.att1(x=x4, g=d1)
        d1 = torch.cat((x4, d1), dim=1)
        d1 = self.d1(d1)
        
        # Block 2
        d2 = self.up2(d1)
        x3 = self.att2(x=x3, g=d2)
        d2 = torch.cat((x3, d2), dim=1)
        d2 = self.d2(d2)
        # [NEW] Capture Intermediate Output (1/4 resolution)
        out3 = self.ds_out3(d2)

        # Block 3
        d3 = self.up3(d2)
        x2 = self.att3(x=x2, g=d3)
        d3 = torch.cat((x2, d3), dim=1)
        d3 = self.d3(d3)
        # [NEW] Capture Intermediate Output (1/2 resolution)
        out2 = self.ds_out2(d3)
        
        # Block 4
        d4 = self.up4(d3)
        x1 = self.att4(x=x1, g=d4)
        d4 = torch.cat((x1, d4), dim=1)
        d4 = self.d4(d4)
        
        # Final Output
        out_final = self.out(d4)
        
        # [NEW] Return List: [Final, Middle, Coarse]
        # We must upsample the smaller maps to match the input size for Loss calculation
        return [
            out_final,
            F.interpolate(out2, scale_factor=2, mode='bilinear', align_corners=True),
            F.interpolate(out3, scale_factor=4, mode='bilinear', align_corners=True)
        ]
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class DiceBCELoss(nn.Module):
    def __init__(self, weight=None, size_average=True):
        super(DiceBCELoss, self).__init__()

    def forward(self, inputs, targets, smooth=1):
        # inputs are LOGITS (raw output from model, no sigmoid yet)
        
        # --- 1. BCE Loss (Must use logits for stability in AMP) ---
        # binary_cross_entropy_with_logits includes the sigmoid internally
        bce_loss = F.binary_cross_entropy_with_logits(inputs, targets, reduction='mean')
        
        # --- 2. Dice Loss (Needs probabilities 0-1) ---
        # We manually apply sigmoid just for the Dice calculation
        inputs_prob = torch.sigmoid(inputs)
        
        # Flatten for Dice math
        inputs_flat = inputs_prob.view(-1)
        targets_flat = targets.view(-1)
        
        intersection = (inputs_flat * targets_flat).sum()                            
        dice_loss = 1 - (2.*intersection + smooth)/(inputs_flat.sum() + targets_flat.sum() + smooth)  
        
        return bce_loss + dice_loss
```

```python
import os
import cv2
import csv
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import albumentations as A 
from albumentations.pytorch import ToTensorV2
from tqdm import tqdm 

# --- IMPORT YOUR MODEL & LOSS ---
# Ensure model.py and loss.py are in the same directory
# from model import ResAttUNet 
# from loss import DiceBCELoss 

# ==========================================
# 1. CONFIGURATION
# ==========================================
LEARNING_RATE = 1e-4
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
BATCH_SIZE = 4 
NUM_EPOCHS = 50
NUM_WORKERS = 2
# AIRS images are large. Crop them to 512x512 or 256x256 to fit in GPU RAM.
IMAGE_HEIGHT = 512 
IMAGE_WIDTH = 512
PIN_MEMORY = True

# Update these paths to match the AIRS structure
BASE_DIR = "/kaggle/input/aerialimageryforroofsegmentation"
TRAIN_IMG_DIR = os.path.join(BASE_DIR, "train/image")
TRAIN_LABEL_DIR = os.path.join(BASE_DIR, "train/label")

VAL_IMG_DIR = os.path.join(BASE_DIR, "val/image")
VAL_LABEL_DIR = os.path.join(BASE_DIR, "val/label")

CSV_FILE = "training_log.csv" 

# ==========================================
# 2. DATASET CLASS (AIRS SPECIFIC)
# ==========================================
class AIRSDataset(Dataset):
    def __init__(self, img_dir, label_dir, transform=None):
        self.img_dir = img_dir
        self.label_dir = label_dir
        self.transform = transform
        # List all TIF images
        self.images = [f for f in os.listdir(img_dir) if f.lower().endswith('.tif')]

    def __len__(self):
        return len(self.images)

    def __getitem__(self, index):
        img_filename = self.images[index]
        img_path = os.path.join(self.img_dir, img_filename)
        
        # AIRS usually keeps the same filename for image and label
        label_path = os.path.join(self.label_dir, img_filename)

        # 1. Load Image
        # Use IMREAD_UNCHANGED to support potential high-bit depth or 4-channel TIFs
        image = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)
        
        # Convert to RGB (OpenCV loads BGR by default)
        if image is None:
             raise FileNotFoundError(f"Image not found: {img_path}")
             
        # Handle 3-channel or 4-channel images
        if len(image.shape) == 2:
            image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
        elif image.shape[2] == 4:
            image = cv2.cvtColor(image, cv2.COLOR_BGRA2RGB)
        else:
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # 2. Load Mask
        if not os.path.exists(label_path):
             raise FileNotFoundError(f"Label not found: {label_path}")
             
        mask = cv2.imread(label_path, cv2.IMREAD_GRAYSCALE)
        
        # 3. Normalize Mask (Critical Step)
        # AIRS masks: 0 = Background, 255 (or 1) = Roof. 
        # We need strictly 0.0 and 1.0 for Binary Cross Entropy.
        mask = mask.astype(np.float32)
        mask[mask > 0] = 1.0 
        
        # 4. Augmentations
        if self.transform is not None:
            augmentations = self.transform(image=image, mask=mask)
            image = augmentations["image"]
            mask = augmentations["mask"]

        return image, mask

# ==========================================
# 3. DST STITCHER (Reuse from previous code)
# ==========================================
def make_mosaic(images, masks):
    """Stitches 4 images into 1 (DST Stitcher)."""
    batch_size = images.shape[0]
    if batch_size < 4: return images, masks 

    stitched_imgs = []
    stitched_masks = []

    for i in range(0, batch_size - 3, 4):
        # Resize to half size
        img1 = F.interpolate(images[i].unsqueeze(0), scale_factor=0.5)
        img2 = F.interpolate(images[i+1].unsqueeze(0), scale_factor=0.5)
        img3 = F.interpolate(images[i+2].unsqueeze(0), scale_factor=0.5)
        img4 = F.interpolate(images[i+3].unsqueeze(0), scale_factor=0.5)

        mask1 = F.interpolate(masks[i].unsqueeze(0).unsqueeze(0), scale_factor=0.5)
        mask2 = F.interpolate(masks[i+1].unsqueeze(0).unsqueeze(0), scale_factor=0.5)
        mask3 = F.interpolate(masks[i+2].unsqueeze(0).unsqueeze(0), scale_factor=0.5)
        mask4 = F.interpolate(masks[i+3].unsqueeze(0).unsqueeze(0), scale_factor=0.5)

        # Stitch
        top = torch.cat([img1, img2], dim=3)    
        bot = torch.cat([img3, img4], dim=3)
        final_img = torch.cat([top, bot], dim=2) 

        top_m = torch.cat([mask1, mask2], dim=3)
        bot_m = torch.cat([mask3, mask4], dim=3)
        final_mask = torch.cat([top_m, bot_m], dim=2).squeeze(1)

        stitched_imgs.append(final_img)
        stitched_masks.append(final_mask)

    if len(stitched_imgs) == 0: return images, masks
    return torch.cat(stitched_imgs, dim=0), torch.cat(stitched_masks, dim=0)

# ==========================================
# 4. TRAINING UTILS
# ==========================================
def get_loaders(train_img_dir, train_label_dir, val_img_dir, val_label_dir, 
                batch_size, train_transform, val_transform, num_workers=4, pin_memory=True):
    
    train_ds = AIRSDataset(
        img_dir=train_img_dir, 
        label_dir=train_label_dir, 
        transform=train_transform
    )
    
    train_loader = DataLoader(
        train_ds, 
        batch_size=batch_size, 
        num_workers=num_workers, 
        pin_memory=pin_memory, 
        shuffle=True,
    )

    val_ds = AIRSDataset(
        img_dir=val_img_dir, 
        label_dir=val_label_dir, 
        transform=val_transform
    )
    
    val_loader = DataLoader(
        val_ds, 
        batch_size=batch_size, 
        num_workers=num_workers, 
        pin_memory=pin_memory, 
        shuffle=False,
    )

    return train_loader, val_loader

def evaluate(loader, model, loss_fn, device="cuda"):
    model.eval()
    val_loss = 0
    dice_score = 0
    iou_score = 0
    precision = 0
    recall = 0
    num_correct = 0
    num_pixels = 0
    
    loop = tqdm(loader, desc="Validating", leave=False)

    with torch.no_grad():
        for x, y in loop:
            x = x.to(device)
            y = y.float().to(device).unsqueeze(1) 
            
            outputs = model(x)
            final_pred = outputs[0]
            
            loss = loss_fn(final_pred, y)
            val_loss += loss.item()
            
            preds = torch.sigmoid(final_pred)
            preds = (preds > 0.5).float() 
            
            TP = (preds * y).sum()
            FP = (preds * (1 - y)).sum()
            FN = ((1 - preds) * y).sum()
            
            smooth = 1e-8
            dice_score += (2 * TP) / (2 * TP + FP + FN + smooth)
            iou_score += TP / (TP + FP + FN + smooth)
            precision += TP / (TP + FP + smooth)
            recall += TP / (TP + FN + smooth)
            
            num_correct += (preds == y).sum().item()
            num_pixels += torch.numel(preds)

    length = len(loader)
    return {
        "val_loss": val_loss / length,
        "accuracy": (num_correct / num_pixels) * 100,
        "dice": (dice_score / length).item(),
        "iou": (iou_score / length).item(),
        "precision": (precision / length).item(),
        "recall": (recall / length).item()
    }

def train_fn(loader, model, optimizer, loss_fn, scaler):
    loop = tqdm(loader, desc="Training")
    USE_MOSAIC = False 
    running_loss = 0.0

    for batch_idx, (data, targets) in enumerate(loop):
        data = data.to(DEVICE)
        targets = targets.float().to(DEVICE) 

        if USE_MOSAIC:
            data, targets = make_mosaic(data, targets)
            USE_MOSAIC = False 

        with torch.amp.autocast('cuda'): 
            outputs = model(data) 
            final_pred, mid_pred, coarse_pred = outputs[0], outputs[1], outputs[2]
            
            targets_unsqueezed = targets.unsqueeze(1)
            
            loss_1 = loss_fn(final_pred, targets_unsqueezed)
            loss_2 = loss_fn(mid_pred, targets_unsqueezed)
            loss_3 = loss_fn(coarse_pred, targets_unsqueezed)
            
            total_loss = loss_1 + (0.5 * loss_2) + (0.4 * loss_3)
            
            if torch.rand(1).item() < 0.2: 
                USE_MOSAIC = True

        optimizer.zero_grad()
        scaler.scale(total_loss).backward()
        scaler.step(optimizer)
        scaler.update()

        running_loss += total_loss.item()
        loop.set_postfix(loss=total_loss.item())
        
    return running_loss / len(loader)

# ==========================================
# 5. MAIN EXECUTION
# ==========================================
def main():
    # --- TRANSFORMS ---
    # AIRS images are large, so we resize them to a manageable size (512x512)
    train_transform = A.Compose([
        A.Resize(height=IMAGE_HEIGHT, width=IMAGE_WIDTH),
        A.Rotate(limit=35, p=1.0),
        A.HorizontalFlip(p=0.5),
        A.VerticalFlip(p=0.1),
        A.Normalize(mean=[0.0, 0.0, 0.0], std=[1.0, 1.0, 1.0], max_pixel_value=255.0),
        ToTensorV2(),
    ])

    val_transform = A.Compose([
        A.Resize(height=IMAGE_HEIGHT, width=IMAGE_WIDTH),
        A.Normalize(mean=[0.0, 0.0, 0.0], std=[1.0, 1.0, 1.0], max_pixel_value=255.0),
        ToTensorV2(),
    ])

    # --- INITIALIZE ---
    model = ResAttUNet(in_channels=3, out_channels=1).to(DEVICE)
    loss_fn = DiceBCELoss() 
    optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
    scaler = torch.amp.GradScaler('cuda')

    train_loader, val_loader = get_loaders(
        TRAIN_IMG_DIR, TRAIN_LABEL_DIR,
        VAL_IMG_DIR, VAL_LABEL_DIR,
        BATCH_SIZE, train_transform, val_transform, NUM_WORKERS, PIN_MEMORY
    )

    # --- CSV SETUP ---
    header = ['epoch', 'train_loss', 'val_loss', 'accuracy', 'dice', 'iou', 'precision', 'recall']
    with open(CSV_FILE, 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(header)

    best_dice_score = 0.0

    print(f"Starting training on {DEVICE}...")

    # --- EPOCH LOOP ---
    for epoch in range(NUM_EPOCHS):
        print(f"\nEpoch {epoch+1}/{NUM_EPOCHS}")
        
        avg_train_loss = train_fn(train_loader, model, optimizer, loss_fn, scaler)
        metrics = evaluate(val_loader, model, loss_fn, DEVICE)
        
        print(f"Train Loss: {avg_train_loss:.4f} | Val Loss: {metrics['val_loss']:.4f}")
        print(f"Dice: {metrics['dice']:.4f} | IoU: {metrics['iou']:.4f} | Acc: {metrics['accuracy']:.2f}%")
        
        if metrics['dice'] > best_dice_score:
            best_dice_score = metrics['dice']
            checkpoint = {
                "state_dict": model.state_dict(),
                "optimizer": optimizer.state_dict(),
                "epoch": epoch,
                "best_dice": best_dice_score,
            }
            torch.save(checkpoint, "best_checkpoint.pth.tar")
            print(f"--> New best model saved! (Dice: {best_dice_score:.4f})")
            
        with open(CSV_FILE, 'a', newline='') as f:
            writer = csv.writer(f)
            writer.writerow([
                epoch + 1, avg_train_loss, metrics['val_loss'], metrics['accuracy'], 
                metrics['dice'], metrics['iou'], metrics['precision'], metrics['recall']
            ])

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- CONFIGURATION ---
CSV_FILE = "/kaggle/working/training_log.csv"
OUTPUT_IMAGE = "metrics_plot.png"

def plot_training_metrics(csv_file, output_file):
    # 1. Load Data
    try:
        df = pd.read_csv(csv_file)
    except FileNotFoundError:
        print(f"Error: Could not find {csv_file}. Make sure you have run the training script first.")
        return

    # 2. Set "Aesthetic" Style (Seaborn)
    sns.set_theme(style="whitegrid", context="paper", font_scale=1.2)
    
    # Create a figure with a 2x3 grid
    fig, axes = plt.subplots(2, 3, figsize=(18, 10))
    fig.suptitle('Training & Validation Metrics', fontsize=20, fontweight='bold', y=0.98)

    # --- PLOT 1: LOSS (Train vs Val) ---
    sns.lineplot(ax=axes[0, 0], data=df, x='epoch', y='train_loss', label='Train Loss', 
                 linewidth=2.5, color='#E74C3C', marker='o')
    sns.lineplot(ax=axes[0, 0], data=df, x='epoch', y='val_loss', label='Val Loss', 
                 linewidth=2.5, color='#2E86C1', marker='o')
    axes[0, 0].set_title('Loss Curves', fontweight='bold')
    axes[0, 0].set_ylabel('Loss')
    axes[0, 0].legend()

    # --- PLOT 2: DICE & IoU (The most important segmentation metrics) ---
    sns.lineplot(ax=axes[0, 1], data=df, x='epoch', y='dice', label='Dice Score', 
                 linewidth=2.5, color='#27AE60', marker='s')
    sns.lineplot(ax=axes[0, 1], data=df, x='epoch', y='iou', label='IoU Score', 
                 linewidth=2.5, color='#F39C12', marker='s')
    axes[0, 1].set_title('Segmentation Performance (Dice & IoU)', fontweight='bold')
    axes[0, 1].set_ylabel('Score (0-1)')
    axes[0, 1].legend()

    # --- PLOT 3: ACCURACY ---
    sns.lineplot(ax=axes[0, 2], data=df, x='epoch', y='accuracy', 
                 linewidth=2.5, color='#8E44AD', marker='D')
    axes[0, 2].set_title('Pixel Accuracy (%)', fontweight='bold')
    axes[0, 2].set_ylabel('Accuracy (%)')

    # --- PLOT 4: PRECISION vs RECALL ---
    sns.lineplot(ax=axes[1, 0], data=df, x='epoch', y='precision', label='Precision', 
                 linewidth=2.5, color='#16A085', marker='^')
    sns.lineplot(ax=axes[1, 0], data=df, x='epoch', y='recall', label='Recall', 
                 linewidth=2.5, color='#C0392B', marker='v')
    axes[1, 0].set_title('Precision vs Recall', fontweight='bold')
    axes[1, 0].set_ylabel('Score (0-1)')
    axes[1, 0].legend()

    # --- PLOT 5: DICE vs LOSS (Correlation check) ---
    # Helps see if Loss going down actually improves Dice
    ax2 = axes[1, 1].twinx()
    sns.lineplot(ax=axes[1, 1], data=df, x='epoch', y='val_loss', color='#2E86C1', 
                 label='Val Loss', legend=False, linestyle='--')
    sns.lineplot(ax=ax2, data=df, x='epoch', y='dice', color='#27AE60', 
                 label='Dice', legend=False, linewidth=2.5)
    
    axes[1, 1].set_title('Correlation: Val Loss (Blue) vs Dice (Green)', fontweight='bold')
    axes[1, 1].set_ylabel('Val Loss')
    ax2.set_ylabel('Dice Score')
    
    # --- PLOT 6: DATA TABLE (Summary) ---
    axes[1, 2].axis('off')
    # Create a concise table of the best epoch
    best_epoch_idx = df['dice'].idxmax()
    best_row = df.iloc[best_epoch_idx]
    
    table_text = [
        ["Metric", "Best Value"],
        ["Best Epoch", int(best_row['epoch'])],
        ["Dice Score", f"{best_row['dice']:.4f}"],
        ["IoU Score", f"{best_row['iou']:.4f}"],
        ["Precision", f"{best_row['precision']:.4f}"],
        ["Recall", f"{best_row['recall']:.4f}"],
        ["Val Loss", f"{best_row['val_loss']:.4f}"],
    ]
    
    table = axes[1, 2].table(cellText=table_text, loc='center', cellLoc='center', 
                             colWidths=[0.4, 0.4])
    table.auto_set_font_size(False)
    table.set_fontsize(12)
    table.scale(1, 2) # Make rows taller
    axes[1, 2].set_title(f"Best Model Summary (Epoch {int(best_row['epoch'])})", fontweight='bold')

    # Final Layout Adjustments
    plt.tight_layout()
    plt.subplots_adjust(top=0.92) # Make space for main title
    
    # Save
    plt.savefig(output_file, dpi=300, bbox_inches='tight')
    print(f"✅ Plots saved successfully to {output_file}")
    plt.show()

if __name__ == "__main__":
    plot_training_metrics(CSV_FILE, OUTPUT_IMAGE)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F  # Added missing import
import numpy as np
import cv2
import matplotlib.pyplot as plt
import albumentations as A
from albumentations.pytorch import ToTensorV2

# --- 2. CONFIGURATION ---
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
CHECKPOINT_PATH = "/kaggle/working/best_checkpoint.pth.tar"
# Replace with a valid image path from your dataset
IMAGE_PATH = "/kaggle/input/aerialimageryforroofsegmentation/train/image/christchurch_1000.tif" 
INPUT_SIZE = 256 

# --- 3. INFERENCE UTILS ---
def load_model(checkpoint_path):
    print(f"Loading model from {checkpoint_path}...")
    model = ResAttUNet(in_channels=3, out_channels=1).to(DEVICE)
    checkpoint = torch.load(checkpoint_path, map_location=DEVICE)
    model.load_state_dict(checkpoint["state_dict"])
    model.eval()
    return model

def predict_single_image(model, image_path):
    # 1. Read Image
    original_img = cv2.imread(image_path)
    if original_img is None:
        raise ValueError(f"Image not found at {image_path}")
    
    original_rgb = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)
    
    # 2. Preprocess
    transform = A.Compose([
        A.Resize(height=INPUT_SIZE, width=INPUT_SIZE),
        A.Normalize(mean=[0.0, 0.0, 0.0], std=[1.0, 1.0, 1.0], max_pixel_value=255.0),
        ToTensorV2(),
    ])
    
    augmented = transform(image=original_rgb)
    img_tensor = augmented["image"].unsqueeze(0).to(DEVICE)
    
    # 3. Inference
    with torch.no_grad():
        outputs = model(img_tensor) # This returns a list: [final, mid, coarse]
        
        # --- FIX IS HERE: Take index 0 ---
        logits = outputs[0] 
        
        probs = torch.sigmoid(logits)
        mask = (probs > 0.5).float()
    
    # 4. Post-processing
    mask_np = mask.squeeze().cpu().numpy()
    original_h, original_w = original_rgb.shape[:2]
    mask_resized = cv2.resize(mask_np, (original_w, original_h), interpolation=cv2.INTER_NEAREST)
    
    return original_rgb, mask_resized

def visualize_result(image, mask):
    mask_colored = np.zeros_like(image)
    mask_colored[:, :, 0] = 255 # Red
    mask_indices = mask > 0
    overlay = image.copy()
    overlay[mask_indices] = cv2.addWeighted(image[mask_indices], 0.5, mask_colored[mask_indices], 0.5, 0)

    fig, ax = plt.subplots(1, 3, figsize=(15, 5))
    ax[0].imshow(image); ax[0].set_title("Original Satellite Image"); ax[0].axis("off")
    ax[1].imshow(mask, cmap="gray"); ax[1].set_title("Predicted Segmentation Mask"); ax[1].axis("off")
    ax[2].imshow(overlay); ax[2].set_title("Overlay (Red = Rooftop)"); ax[2].axis("off")
    plt.tight_layout()
    plt.show()

# --- MAIN EXECUTION ---
if __name__ == "__main__":
    model = load_model(CHECKPOINT_PATH)
    try:
        img, pred_mask = predict_single_image(model, IMAGE_PATH)
        visualize_result(img, pred_mask)
        print("Inference Complete!")
    except Exception as e:
        print(f"Error: {e}")
```
